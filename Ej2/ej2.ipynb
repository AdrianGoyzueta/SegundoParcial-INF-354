```python
import pandas as pd


def red_neuronal(S: list, W: list, ys:list, t_a: float):
  if len(S) != len(W):
    raise ValueError("La cantidad de neuronas en S y W debe ser la misma")

  if len(ys) != len(S[0]):
    raise ValueError("La cantidad de salidas esperadas debe ser la misma que la cantidad de salidas de la red")

  df = {}

  for i, s in enumerate(S, start=1):
    df[f"s{i}"] = s

  for i, w in enumerate(W, start=1):
    df[f"w{i}"] = [w] * len(S[0])

  df["yc"] = [None] * len(S[0])
  df["ys"] = ys

  for i, w in enumerate(W, start=1):
    df[f"wn{i}"] = [0] * len(S[0])

  for i in range(len(S[0])):
    if i != 0:
      for j in range(1, len(W)+1):
        df[f"w{j}"][i] = df[f"w{j}"][i-1] + df[f"wn{j}"][i-1]
    df["yc"][i] = int(sum(df[f"s{j}"][0] * df[f"w{j}"][0] for j in range(1, len(W)+1)) > 0)
    for j in range(1, len(W)+1):
      df[f"wn{j}"][i] = t_a * df[f"s{j}"][i] * (df["ys"][i] - df["yc"][i])

  return df


S = [[1, 1, 1], [0, 0, 1], [0, 1, 0]]
W = [1, 1, 1]
ys = [0, 0, 0]
t_a = 0.2

# Solo se usa pandas para mostrar los datos de una forma m√°s presentable
pd.DataFrame(red_neuronal(S, W, ys, t_a))
```
